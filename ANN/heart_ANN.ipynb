```python
from torch.utils.data import Dataset, DataLoader
```

```python
class CustomDataset(Dataset):
  def __init__(self, features, labels):
    self.features = features
    self.labels =labels

  def __len__(self):
    return len(self.features)

  def __getitem__(self, index):
    return self.features[index], self.labels[index]

```

```python
train_dataset = CustomDataset(X_train_tensor,y_train_tensor)
test_dataset = CustomDataset(X_test_tensor,y_test_tensor)
```

```python
train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle =True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=32,
    shuffle =True
)
```

Define model

```python
from torch import nn as nn
class Model(nn.Module):
  def __init__(self, num_features):
    super().__init__()

    self.linear1=nn.Linear(num_features,3)
    self.ReLu=nn.ReLU()

    self.linear2=nn.Linear(3,1)
    self.sigmoid=nn.Sigmoid()

  def forward(self,features):
   out=self.linear1(features)
   out=self.ReLu(out)

   out= self.linear2(out)
   out= self.sigmoid(out)
   return out




```

```python
learning_rate = 0.1
epochs = 100
loss_history=[]
```

```python
#create model

model = Model(X_train_tensor.shape[1])

#optimizer
optimizer = torch.optim.SGD( model.parameters(), lr = learning_rate )

#define loop

for epoch in range(epochs):

  for batch_feature,batch_labels in train_loader:
    #forward pass
    y_pred = model(batch_feature)
    #Loss calculate
    loss = loss_function(y_pred,batch_labels.reshape(-1,1))

    #zero gradient
    optimizer.zero_grad()
    #backward pass
    loss.backward()

    #parameter update
    optimizer.step()



  #print loss in each epoch

  print(f" {epoch+1} , Loss = {loss.item()} ")
```

3.Zero Gradient:it is used for clear previous gradient,sothat old error don't prevent new calculation.

## Write Answer 07:
1.In PyTorch, gradients are added to the .grad attribute of parameters whenever loss.backward() is called. If you do not call optimizer.zero_grad() at the start of each training step, the new gradients will be added to the ones calculated in previous steps.

2.A high learning rate can cause:
   a)The model to overshoot the optimal weights.
   b)Loss may oscillate.
   c)Convergence might never happen.

3.A very small learning rate causes:
  a)Slower convergence.
  b)Model takes much longer to reach the minimum loss.

4.We define layers inside __init__ so PyTorch can register their weights and biases as trainable parameters for the optimizer.
If layers were defined inside forward(), new layers would be created on every pass, and the optimizer wouldn’t know which parameters to update
